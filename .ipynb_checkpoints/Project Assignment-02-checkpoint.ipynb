### Machine learning time
**Random forest (regression - supervised)**

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

**Preprocess data so that (i) categorical variables are numeric and (ii) data is scaled so that features have the same level of magnitudes**

```python
# convert categorical variable into dummy/indicator variables before using encoder to make numeric array
df_onehot = pd.get_dummies(df, columns=['Year','Area','Item'])

df_onehot.head()
```

```python
# use onehot encoder on features (predictor variables), which are everything but crop yield
x = df_onehot.loc[:, df_onehot.columns != 'hg/ha_yield']

# identify response variable
y = df['hg/ha_yield']


#x = x.drop(['Year'], axis=1)

x.head()
```

```python
scaler = MinMaxScaler()
x = scaler.fit_transform(x) 
```

**Split features and target data into training and test sets**

```python
x_train, x_test, y_train, y_test = train_test_split(
        x, y, test_size=0.2, random_state=42) #stratification not needed here because we're not working with groups/classes
```

**determine appropriate depth and number of estimators to minimize root mean square error**

```python
depth_range = range(1, 30) #bumped up to 30 because it was just leveling out at 20
depth_scores = []
for d in depth_range:
    rf_regressor = RandomForestRegressor(max_depth = d)
    score = cross_val_score(rf_regressor,
                           x_train,
                           y_train, 
                           cv=5, 
                           scoring='neg_root_mean_squared_error')
    depth_scores.append(score.mean())
```

```python
import matplotlib.pyplot as plt
plt.scatter(depth_range, depth_scores)
plt.xlabel('Value of Max Depth for RF Classifier')
plt.ylabel('RMSE')
```

looks like 25 would be the optimum depth

**Train model using training data and assign predictions using test data**

```python
rf_regressor = RandomForestRegressor(max_depth=25)
rf_regressor.fit(x_train, y_train)
y_pred = rf_regressor.predict(x_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

The RSME is pretty high (greater than 10% of the average crop yield, which is 77053). Let's see if increasing the number of trees from the default of 100 to 200 helps...

```python
rf_regressor = RandomForestRegressor(n_estimators = 200, max_depth=25)
rf_regressor.fit(x_train, y_train)
y_pred = rf_regressor.predict(x_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

The model still isn't performing well, worse in fact, so either other types of regression models should be explored or the data may need to be maniulated further??
